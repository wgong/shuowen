```python
conn = duckdb.connect("shuowen.duckdb")
```

```python
table_name = "zi"
```

```python
DELETE_ZI = False #  True  # 
```

```python
if DELETE_ZI:
    sql_stmt = f"""
            delete from {table_name};
        """
    conn.sql(sql_stmt)
```

```python
imax = 9834 # 3 # 

for i in range(1, imax):
    file_json = Path(f"data/{i}.json")
    if not file_json.exists(): 
        continue 
        
    if i == 1:
        sql_stmt = f"""
            create table IF NOT EXISTS {table_name} as 
            select * from read_json_auto('{file_json}');
        """
    else:
        sql_stmt = f"""
        insert into {table_name} 
        select * from read_json_auto('{file_json}');
        """
    conn.sql(sql_stmt)
```

```python
# extract parts 

sql_stmt = f"""
select 
    id, wordhead,  explanation, 
    regexp_extract(explanation, '.*从(.)(.)聲.*' , 1) as p1,
    regexp_extract(explanation, '.*从(.)(.)聲.*' , 2) as p2

from {table_name} 
where 1=1
    --and wordhead = '祖' 
    and regexp_matches(explanation, '.*从..聲.*' )
"""
zi = conn.sql(sql_stmt).df()
zi
```
